```python
import pandas as pd
import numpy as np
from scipy import stats

def run_task3_ab_testing(df):
    print("=== Starting Task 3: A/B Hypothesis Testing Pipeline ===\n")
    
    # Pre-test data cleaning check to avoid failures on null rows
    df = df.copy()
    df['Margin'] = df['TotalPremium'] - df['TotalClaims']
    df['HasClaim'] = np.where(df['TotalClaims'] > 0, 1, 0)
    
    # -------------------------------------------------------------------------
    # TEST 1: GENDER RISK (KPI: MARGIN)
    # -------------------------------------------------------------------------
    print("--- Test 1: Gender Risk Profiling (KPI: Margin) ---")
    control_gender = df[df['Gender'].str.upper() == 'FEMALE']['Margin']
    test_gender = df[df['Gender'].str.upper() == 'MALE']['Margin']
    
    if len(control_gender) > 0 and len(test_gender) > 0:
        t_stat, p_val_t1 = stats.ttest_ind(test_gender, control_gender, equal_var=False)
        print(f"Control (Female) Size: {len(control_gender)}, Test (Male) Size: {len(test_gender)}")
        print(f"t-statistic: {t_stat:.4f} | p-value: {p_val_t1:.4e}")
        if p_val_t1 < 0.05:
            print("Decision: Reject Null Hypothesis. Risk margin varies significantly by gender.\n")
        else:
            print("Decision: Fail to Reject Null Hypothesis. No significant risk margin difference found.\n")
            
    # -------------------------------------------------------------------------
    # TEST 2: GEOGRAPHIC VARIANCE (KPI: CLAIM FREQUENCY)
    # -------------------------------------------------------------------------
    print("--- Test 2: Geographic Variance (KPI: Claim Frequency) ---")
    # Filter for target provinces
    geo_df = df[df['Province'].str.upper().isin(['NC', 'GP'])]
    
    if len(geo_df) > 0:
        # Create a contingency table: Rows = Provinces, Columns = HasClaim (0 or 1)
        contingency_table = pd.crosstab(geo_df['Province'], geo_df['HasClaim'])
        print("Contingency Table (Province vs Claim Occurrence):")
        print(contingency_table)
        
        chi2_stat, p_val_t2, dof, expected = stats.chi2_contingency(contingency_table)
        print(f"Chi2 Statistic: {chi2_stat:.4f} | p-value: {p_val_t2:.4e}")
        if p_val_t2 < 0.05:
            print("Decision: Reject Null Hypothesis. Claim frequency is strongly dependent on geography.\n")
        else:
            print("Decision: Fail to Reject Null Hypothesis. Geography has no statistical impact on claims.\n")

    # -------------------------------------------------------------------------
    # TEST 3: SECURITY FEATURES IMPACT (KPI: CLAIM SEVERITY)
    # -------------------------------------------------------------------------
    print("--- Test 3: Security Features Impact (KPI: Claim Severity) ---")
    # Severity is only evaluated when an actual financial loss occurred
    claims_subset = df[df['TotalClaims'] > 0]
    
    control_security = claims_subset[claims_subset['TrackingDevice'] == 0]['TotalClaims']
    test_security = claims_subset[claims_subset['TrackingDevice'] == 1]['TotalClaims']
    
    if len(control_security) > 0 and len(test_security) > 0:
        t_stat_t3, p_val_t3 = stats.ttest_ind(test_security, control_security, equal_var=False)
        print(f"Control (No Tracker) Claims: {len(control_security)}, Test (With Tracker) Claims: {len(test_security)}")
        print(f"t-statistic: {t_stat_t3:.4f} | p-value: {p_val_t3:.4e}")
        if p_val_t3 < 0.05:
            print("Decision: Reject Null Hypothesis. Trackers significantly alter claim severity distributions.\n")
        else:
            print("Decision: Fail to Reject Null Hypothesis. Security tracking shows no structural impact on claim severity.\n")

# To execute this utility function within your notebook or main runner pipeline:
# run_task3_ab_testing(your_cleaned_dataframe)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from scipy import stats

def extract_balanced_test_segments(df):
    print("=== Commencing Controlled Data Segmentation Suite ===")
    
    # 1. Isolate the target baseline and comparison categories
    target_data = df[df['Province'].str.upper().isin(['NC', 'GP'])].copy()
    
    # Encode group assignment: Control (NC) = 0, Test (GP) = 1
    target_data['Group_Assignment'] = np.where(target_data['Province'].str.upper() == 'GP', 1, 0)
    
    # 2. Define the confounding attributes to be equalized
    confounders_numeric = ['Cubiccapacity', 'Kilowatts', 'CustomValueEstimate', 'SumInsured', 'ExcessSelected']
    confounders_categorical = ['Gender', 'MaritalStatus']
    
    # Data Cleaning: Handle missing values in confounders defensively
    for col in confounders_numeric:
        target_data[col] = pd.to_numeric(target_data[col], errors='coerce')
        target_data[col].fillna(target_data[col].median(), inplace=True)
        
    for col in confounders_categorical:
        target_data[col] = target_data[col].fillna('Unknown').astype(str)
        
    # One-hot encode categorical confounders
    encoded_cats = pd.get_dummies(target_data[confounders_categorical], drop_first=True)
    X_features = pd.concat([target_data[confounders_numeric], encoded_cats], axis=1)
    y_target = target_data['Group_Assignment']
    
    # 3. Compute Propensity Scores using a logistic regression classifier
    print("Calculating propensity scores to align confounding attributes...")
    propensity_model = LogisticRegression(max_iter=1000, random_state=42)
    propensity_model.fit(X_features, y_target)
    target_data['Propensity_Score'] = propensity_model.predict_proba(X_features)[:, 1]
    
    # Separate groups
    control_pool = target_data[target_data['Group_Assignment'] == 0].sort_values(by='Propensity_Score')
    test_pool = target_data[target_data['Group_Assignment'] == 1].sort_values(by='Propensity_Score')
    
    # 4. Nearest-Neighbor Propensity Matching to extract statistically equivalent cohorts
    print("Executing nearest-neighbor matching algorithm...")
    matched_control_indices = []
    matched_test_indices = []
    
    # Match the smaller group to the larger group without replacement
    smaller_pool = control_pool if len(control_pool) < len(test_pool) else test_pool
    larger_pool = test_pool if len(control_pool) < len(test_pool) else control_pool
    
    larger_pool_scores = larger_pool['Propensity_Score'].to_numpy()
    larger_pool_indices = larger_pool.index.to_numpy()
    
    for idx, row in smaller_pool.iterrows():
        score = row['Propensity_Score']
        # Find index of closest propensity score in the larger pool
        closest_idx = (np.abs(larger_pool_scores - score)).argmin()
        
        matched_control_indices.append(idx if row['Group_Assignment'] == 0 else larger_pool_indices[closest_idx])
        matched_test_indices.append(larger_pool_indices[closest_idx] if row['Group_Assignment'] == 0 else idx)
        
        # Remove matched sample from search array to ensure 1:1 matching without replacement
        larger_pool_scores = np.delete(larger_pool_scores, closest_idx)
        larger_pool_indices = np.delete(larger_pool_indices, closest_idx)
        
        if len(larger_pool_scores) == 0:
            break

    group_a_control = df.loc[matched_control_indices].copy()
    group_b_test = df.loc[matched_test_indices].copy()
    
    print(f"\nSuccessfully Segmented Cohorts!")
    print(f"-> Group A (Control - Baseline NC): {len(group_a_control)} records")
    print(f"-> Group B (Test - Comparison GP): {len(group_b_test)} records")
    
    # 5. Multivariate Validation Balance Check
    print("\n--- Running Balance Checks On Confounders (Goal: p-value > 0.05) ---")
    for col in confounders_numeric:
        stat, p_val = stats.ttest_ind(group_a_control[col], group_b_test[col], equal_var=False)
        status = "PASSED (Balanced)" if p_val > 0.05 else "FAILED (Imbalanced)"
        print(f"Feature '{col}': t-stat={stat:.4f}, p-value={p_val:.4f} -> {status}")
        
    print("========================================================\n")
    return group_a_control, group_b_test

# Execution Command:
# group_a, group_b = extract_balanced_test_segments(cleaned_insurance_dataframe)
```

```python
import pandas as pd
import numpy as np
import logging
from scipy import stats

# Configure logging for audit clarity
logging.basicConfig(level=logging.INFO, format="%(asctime)s [%(levelname)s] %(message)s")
logger = logging.getLogger(__name__)

def run_portfolio_hypothesis_tests(df):
    """
    Executes the Task 3 A/B Hypothesis Testing suite using appropriate 
    statistical models based on KPI type.
    
    Significance Threshold: Alpha = 0.05
    """
    logger.info("Initializing Task 3 Statistical Testing Suite...")
    
    # -------------------------------------------------------------------------
    # PRE-CHECK & RECOVERY FOR DERIVED DATA LABELS
    # -------------------------------------------------------------------------
    test_df = df.copy()
    test_df['Margin'] = test_df['TotalPremium'] - test_df['TotalClaims']
    test_df['HasClaim'] = np.where(test_df['TotalClaims'] > 0, 1, 0)
    
    results_log = {}
    alpha = 0.05

    # -------------------------------------------------------------------------
    # HYPOTHESIS 1: GENDER RISK PROFILING (KPI: Margin - Numerical)
    # -------------------------------------------------------------------------
    logger.info("Executing Hypothesis 1: Gender Risk Profiling...")
    
    # Segment Groups
    group_a_gender = test_df[test_df['Gender'].str.upper() == 'FEMALE']['Margin']
    group_b_gender = test_df[test_df['Gender'].str.upper() == 'MALE']['Margin']
    
    if len(group_a_gender) == 0 or len(group_b_gender) == 0:
        logger.error("Hypothesis 1 skipped: Gender features are missing or incomplete.")
    else:
        # Numerical KPI -> Welch's t-test (handles unequal variances robustly)
        t_stat, p_value_t1 = stats.ttest_ind(group_b_gender, group_a_gender, equal_var=False)
        results_log['Hypothesis_1'] = {'test_stat': t_stat, 'p_value': p_value_t1}
        
        print("\n========================================================")
        print("HYPOTHESIS 1 TEST RESULTS (Gender vs. Underwriting Margin)")
        print("========================================================")
        print(f"Group A (Control - Female) Mean Margin: R{group_a_gender.mean():,.2f} (n={len(group_a_gender)})")
        print(f"Group B (Test - Male) Mean Margin:     R{group_b_gender.mean():,.2f} (n={len(group_b_gender)})")
        print(f"Computed t-statistic:                 {t_stat:.4f}")
        print(f"Recorded p-value:                     {p_value_t1:.4e}")
        
        if p_value_t1 < alpha:
            print(f"Outcome: Reject H0 at alpha={alpha}. Gender is a statistically validated risk driver.")
        else:
            print(f"Outcome: Fail to Reject H0. No structural margin discrepancy between genders.")

    # -------------------------------------------------------------------------
    # HYPOTHESIS 2: GEOGRAPHIC VARIANCE (KPI: Claim Frequency - Categorical)
    # -------------------------------------------------------------------------
    logger.info("Executing Hypothesis 2: Geographic Risk Variance...")
    
    # Segment Groups (Control: Low-Risk NC vs Test: High-Risk GP)
    geo_subset = test_df[test_df['Province'].str.upper().isin(['NC', 'GP'])]
    
    if len(geo_subset) == 0:
        logger.error("Hypothesis 2 skipped: Target Provinces 'NC' or 'GP' missing from records.")
    else:
        # Categorical KPI -> Construct 2x2 Contingency Matrix
        contingency_matrix = pd.crosstab(geo_subset['Province'], geo_subset['HasClaim'])
        
        # Categorical Test -> Chi-Square Test of Independence
        chi2_stat, p_value_t2, dof, expected = stats.chi2_contingency(contingency_matrix)
        results_log['Hypothesis_2'] = {'test_stat': chi2_stat, 'p_value': p_value_t2}
        
        print("\n========================================================")
        print("HYPOTHESIS 2 TEST RESULTS (Geography vs. Claim Frequency)")
        print("========================================================")
        print("Observed Contingency Matrix (Rows=Province, Columns=[No Claim, Claim]):")
        print(contingency_matrix.to_string())
        print(f"Computed Chi-Square Statistic:        {chi2_stat:.4f}")
        print(f"Recorded p-value:                     {p_value_t2:.4e}")
        print(f"Degrees of Freedom (df):              {dof}")
        
        if p_value_t2 < alpha:
            print(f"Outcome: Reject H0 at alpha={alpha}. Location directly impacts claim probability.")
        else:
            print(f"Outcome: Fail to Reject H0. Baseline claim occurrence rates are independent of location.")

    # -------------------------------------------------------------------------
    # HYPOTHESIS 3: SECURITY TECH IMPACT (KPI: Claim Severity - Numerical)
    # -------------------------------------------------------------------------
    logger.info("Executing Hypothesis 3: Security Tech Impact on Severity...")
    
    # Severity is isolated exclusively for cases where an actual loss occurred
    claims_only = test_df[test_df['TotalClaims'] > 0]
    
    # Segment Groups (0 = Untracked, 1 = Tracked)
    group_a_security = claims_only[claims_only['TrackingDevice'] == 0]['TotalClaims']
    group_b_security = claims_only[claims_only['TrackingDevice'] == 1]['TotalClaims']
    
    if len(group_a_security) == 0 or len(group_b_security) == 0:
        logger.warning("Hypothesis 3 skipped: Tracking device tracking markers are insufficient or missing.")
    else:
        # Numerical KPI -> Welch's t-test on actual positive losses
        t_stat_t3, p_value_t3 = stats.ttest_ind(group_b_security, group_a_security, equal_var=False)
        results_log['Hypothesis_3'] = {'test_stat': t_stat_t3, 'p_value': p_value_t3}
        
        print("\n========================================================")
        print("HYPOTHESIS 3 TEST RESULTS (Tracking Tech vs. Claim Severity)")
        print("========================================================")
        print(f"Group A (Control - No Tracker) Mean Claim: R{group_a_security.mean():,.2f} (n={len(group_a_security)})")
        print(f"Group B (Test - Active Tracker) Mean Claim: R{group_b_security.mean():,.2f} (n={len(group_b_security)})")
        print(f"Computed t-statistic:                 {t_stat_t3:.4f}")
        print(f"Recorded p-value:                     {p_value_t3:.4e}")
        
        if p_value_t3 < alpha:
            print(f"Outcome: Reject H0 at alpha={alpha}. Active security features significantly mitigate claim size.")
        else:
            print(f"Outcome: Fail to Reject H0. Tracking technology has no noticeable effect on individual loss magnitude.")
            
    print("\n========================================================\n")
    return results_log

# Usage Execution Template:
# hypothesis_p_values = run_portfolio_hypothesis_tests(cleaned_insurance_dataframe)
```
