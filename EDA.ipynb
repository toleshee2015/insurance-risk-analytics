```python
missing_counts = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Values': missing_counts,
    'Percentage (%)': missing_percent
}).sort_values(by='Percentage (%)', ascending=False)

print(missing_df.head(20))
```

```python
import matplotlib.pyplot as plt

missing_df[missing_df['Missing Values'] > 0]['Percentage (%)'].plot(
    kind='bar',
    figsize=(10,4),
    title="Missing Values Percentage by Column"
)
plt.ylabel("Percentage")
plt.show()
```

```python
def classify_missing(pct):
    if pct == 0:
        return "No missing"
    elif pct < 5:
        return "Low"
    elif pct < 20:
        return "Moderate"
    else:
        return "High"

missing_df['Severity'] = missing_df['Percentage (%)'].apply(classify_missing)

print(missing_df.head(20))
```

```python
df = df.dropna(subset=['TransactionMonth'])  # example for critical fields
```

```python
# numeric imputation
num_cols = df.select_dtypes(include=['number']).columns
for col in num_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].median())

# categorical imputation (future-proof)
cat_cols = df.select_dtypes(include=['object', 'string']).columns
for col in cat_cols:
    if df[col].isnull().any() and not df[col].mode(dropna=True).empty:
        df[col] = df[col].fillna(df[col].mode(dropna=True)[0])
```

```python
Q1 = df['TotalPremium'].quantile(0.25)
Q3 = df['TotalPremium'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_premium = df[(df['TotalPremium'] < lower) | (df['TotalPremium'] > upper)]

print("Premium outliers:", len(outliers_premium))
```

```python
Q1 = df['TotalClaims'].quantile(0.25)
Q3 = df['TotalClaims'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_claims = df[(df['TotalClaims'] < lower) | (df['TotalClaims'] > upper)]

print("Claims outliers:", len(outliers_claims))
```
